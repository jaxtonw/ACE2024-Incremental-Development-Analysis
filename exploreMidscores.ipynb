## Structure of "Starter Code" Can Significantly Impact MID Scores

Note the first state is the only difference between the two `FILE_STATES_*`. 

The first has the following:
```
    val = <FILLIN> # UPDATE THIS 
    val *= <FILLIN> # UPDATE THIS
```

While the second has the following:
```
    val = 0 # UPDATE THIS 
    val *= 0 # UPDATE THIS
```

As can be seen below, one of these fake projects is deemed as "Incremental" while the second is not. 

Additionally, the length of the starter code is included in the computation of a MID score as if the student wrote all of the code. It may be beneficial to be able to 'prime' a MID score computation with starter code, such that the starter code does not count towards a students MID score computation.

````python
print(calculate_mid(FILE_STATES_1))
print(calculate_mid(FILE_STATES_2))
````
